```python
X=df[['Age','Annual_Income','Claim_History','Risk_Score','Fraud_Label','Premium_Amount','Policy_Type','Gender']]
```

```python
X
```

```python
Y=df[['Claim_Amount']]
```

```python
Y
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
le=LabelEncoder()

X['Risk_Score']=le.fit_transform(X['Risk_Score'])

```

```python
ohe = OneHotEncoder(sparse_output=False)  
policy_encoded = ohe.fit_transform(X[['Policy_Type']])


policy_df = pd.DataFrame(policy_encoded, columns=ohe.get_feature_names_out(['Policy_Type'])).astype(int)

X = X.drop(columns=['Policy_Type']).reset_index(drop=True)
X = pd.concat([X, policy_df], axis=1)
```

```python
X
```

```python
gender_encoded=ohe.fit_transform(X[['Gender']])
gender_df=pd.DataFrame(gender_encoded,columns=ohe.get_feature_names_out(['Gender'])).astype(int)
X=X.drop(columns=['Gender']).reset_index(drop=True)
X=pd.concat([X,gender_df],axis=1)
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import MinMaxScaler
se=MinMaxScaler()
X_scaled=se.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
X_train,X_test,Y_train,Y_test=train_test_split(X_scaled,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)

model.fit(X_train, Y_train)
```

```python
ypred_train=model.predict(X_train)
ypred_test=model.predict(X_test)
print(r2_score(ypred_test,Y_test))
print(r2_score(ypred_train,Y_train))
```
