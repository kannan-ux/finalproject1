```python
### correct one 


import pandas as pd
df=pd.read_csv(r"C:\Users\sakth\Downloads\synthetic_insurance_data_no_encoding.csv")
import warnings
warnings.filterwarnings("ignore")
```

```python
df
```

```python
X=df[['Age','Annual_Income','Claim_History','Claim_Amount','Fraud_Label','Premium_Amount','Policy_Type','Gender']]
```

```python
X
```

```python
Y=df[['Risk_Score']]
```

```python
Y
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)  
policy_encoded = ohe.fit_transform(X[['Policy_Type']])
policy_df = pd.DataFrame(policy_encoded, columns=ohe.get_feature_names_out(['Policy_Type'])).astype(int)
X = X.drop(columns=['Policy_Type']).reset_index(drop=True)
X = pd.concat([X, policy_df], axis=1)
gender_encoded=ohe.fit_transform(X[['Gender']])
gender_df=pd.DataFrame(gender_encoded,columns=ohe.get_feature_names_out(['Gender'])).astype(int)
X=X.drop(columns=['Gender']).reset_index(drop=True)
X=pd.concat([X,gender_df],axis=1)
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import MinMaxScaler
se=MinMaxScaler()
X_scaled=se.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,classification_report
X_train,X_test,Y_train,Y_test=train_test_split(X_scaled,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
modelclass = RandomForestClassifier(n_estimators=100, random_state=42)

modelclass.fit(X_train, Y_train)
```

```python
ypred=modelclass.predict(X_test)
full_report=classification_report(ypred,Y_test)

```
