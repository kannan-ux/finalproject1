```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("punkt_tab")
```

```python
stop_words=set(stopwords.words("english"))
lemma=WordNetLemmatizer()
import re
def clean(text):
  if text is None:
    return ''
  text=re.sub(r'https\S+','',text)
  text=re.sub('[^a-zA-Z]'," ",text)
  text=text.lower()
  text=word_tokenize(text)
  text=[lemma.lemmatize(word=w,pos='v')for w in text if w not in stop_words if len(w)>2]
  text=' '.join(text)
  return text
df['cleantext']=df['Policy_Text_EN'].apply(clean)
df['cleansummary']=df['Summarized_Text_EN'].apply(clean)
```

```python
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer
import pandas as pd
from google.colab import files
model_name = "facebook/m2m100_418M"
tokenizer = M2M100Tokenizer.from_pretrained(model_name)
model = M2M100ForConditionalGeneration.from_pretrained(model_name)
def translate_text(text, src_lang, tgt_lang):
    tokenizer.src_lang = src_lang
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    translated_tokens = model.generate(**inputs, forced_bos_token_id=tokenizer.get_lang_id(tgt_lang))
    return tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
df2 = pd.read_csv(r"C:\Users\sakth\Documents\Desktop\texting.csv", encoding="latin")
target_languages = input("Enter target languages (fr for French, ta for Tamil, ja for Japanese, separated by commas): ")
selected_languages = [lang.strip() for lang in target_languages.split(",")]
valid_languages = {"fr", "ta", "ja"}
if all(lang in valid_languages for lang in selected_languages):
    for target_lang in selected_languages:
        df2[f"cleantext_{target_lang}"] = df2["cleantext"].apply(lambda x: translate_text(str(x), "en", target_lang))
        df2[f"cleansummary_{target_lang}"] = df2["cleansummary"].apply(lambda x: translate_text(str(x), "en", target_lang))
    output_file = "/content/translated_texts.csv"
    df2.to_csv(output_file, index=False, encoding="utf-8")
    files.download(output_file)
    print(f"✅ Translation complete! File saved as {output_file}")

else:
    print("❌ Invalid language choice. Please enter 'fr', 'ta', or 'ja'.")

```
